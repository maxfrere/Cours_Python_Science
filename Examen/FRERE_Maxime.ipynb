```python
import numpy as np 

# Question 1 :
tab_billet_1 = np.array([50,20,10,5])
somme_a_rendre_1 = 285

# Algorithme de résolution
tab_rendu = np.array([0,0,0,0])
reste = somme_a_rendre_1
for idx, value in enumerate(tab_billet_1):
    quotient = reste//value
    reste = reste%value
    tab_rendu[idx] = quotient

# Affichage du résultat
print("===> Question 1 <===")
print(f"Sur {somme_a_rendre_1}, il vous faudra rendre :")
for idx_bill,value_bill in enumerate(tab_billet_1):
    print(f"{tab_rendu[idx_bill]} billet(s)/pièce(s) de {value_bill} €")
    
# Question 2 :
tab_billet_2 = np.array([50,20,10,5,2,1,0.5,0.20,0.10,0.05,0.02,0.01])
somme_a_rendre_2 = 239.52

# Algorithme de résolution
tab_rendu_2 = np.zeros(len(tab_billet_2))
reste = somme_a_rendre_1
for idx, value in enumerate(tab_billet_2):
    quotient = reste//value
    reste = reste%value
    tab_rendu_2[idx] = quotient

# Affichage du résultat
print("===> Question 2 <===")
print(f"Sur {somme_a_rendre_2}, il vous faudra rendre :")
for idx_bill,value_bill in enumerate(tab_billet_2):
    print(f"{tab_rendu_2[idx_bill]} billet(s)/pièce(s) de {value_bill} €")
    

# Question 3 :
tab_billet_3 = np.array([50,20,10,2,1,0.01])
somme_a_rendre_3 = 239.52

# Algorithme de résolution
def rendu_sur_sommes(somme,tableau_billet):
    tab_rendu = np.zeros(len(tableau_billet))
    reste = somme
    for idx, value in enumerate(tableau_billet):
        quotient = reste//value
        reste = reste%value
        tab_rendu[idx] = quotient
    return tab_rendu

tab_rendu_3 = rendu_sur_sommes(somme_a_rendre_3, tab_billet_3)

# Affichage du résultat
print("===> Question 3 <===")
print(f"Sur {somme_a_rendre_3}, il vous faudra rendre :")
for idx_bill,value_bill in enumerate(tab_billet_3):
    print(f"{tab_rendu_3[idx_bill]} billet(s)/pièce(s) de {value_bill} €")
    

# Question 4 :    
tab_billet_3 = np.array([50,20,10,5,2,1])
stock_billet = np.array([2,5,10,10,3,10])
somme_a_rendre_3 = 353

# Algorithme de résolution
tab_rendu_4 = rendu_sur_sommes(somme_a_rendre_3,tab_billet_3)
diff_tab = stock_billet - tab_rendu_4

print("===> Question 4 <===")
print(f"Le nouveau stock est : \n {diff_tab}")
#Il faut compenser avec d'autre billet pour ne jamais avoir de stoque négatif
new_stock = np.zeros(len(tab_billet_3))
temp_somme = 0
for idx, value in enumerate(diff_tab):
    if value < 0:
        temp_somme += -1*value*tab_billet_3[idx]
        new_stock[idx] = 0
        print(f"Il manque {temp_somme} € de billet(s)/pièce(s) de {tab_billet_3[idx]} €")
    elif value > 0 and temp_somme!=0:
        partial = value*tab_billet_3[idx]
        partial = temp_somme-partial
        if partial > 0:
            print(f"{partial-temp_somme} remboursé par {value} billet(s)/pièce(s) de {tab_billet_3[idx]}")
            temp_somme = partial
            new_stock[idx] = 0
        elif partial < 0:
            quotient = temp_somme // tab_billet_3[idx]
            print(f"{quotient*tab_billet_3[idx]} remboursé par {quotient} billet(s)/pièce(s) de {tab_billet_3[idx]}")
            temp_somme = temp_somme % tab_billet_3[idx]
            new_stock[idx] = value-quotient
        else:
            print(f"{partial-temp_somme} remboursé par {value} billet(s)/pièce(s) de {tab_billet_3[idx]}")
            temp_somme = 0
            new_stock[idx] = 0          
    else:
        new_stock[idx] = diff_tab[idx]
        print(f"Le stock de billet(s)/pièce(s) des {tab_billet_3[idx]} € est inchangé")

print(f"Le stock ré-équilibré est : \n {new_stock} avec un reste de {temp_somme} €")     
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Question 1 :
x_array = np.arange(0.1,100,0.1)

# Question 2 :
f_array = np.exp(3-(100/x_array**2))

# Question 3 :
plt.figure() #Instantier l'objet figure 

plt.plot(x_array,f_array) #Tracer y en fonction de x

plt.title("Graphe question 3") #Mettre un titre (optionel)
plt.xlabel("x") #Nommer l'axe des x (optionel)
plt.ylabel("f(x)") #Nommer  l'axe des y (optionel)
plt.xlim(0,100) #fixer la limite d'affichage sur x (optionel)

plt.show() #Afficher la résultante

# Question 4 :
y_array = np.arange(0.1,100.2,0.2)

# Question 5 :
g_array = 20-(y_array/8)+2*np.cos((np.pi/6)*y_array)

# Question 6 :
plt.figure() #Instantier l'objet figure 

plt.plot(y_array,g_array) #Tracer y en fonction de x

plt.title("Graphe question 6") #Mettre un titre (optionel)
plt.xlabel("x") #Nommer l'axe des x (optionel)
plt.ylabel("f(x)") #Nommer  l'axe des y (optionel)
plt.xlim(0,100) #fixer la limite d'affichage sur x (optionel)

plt.show() #Afficher la résultante

# Question 7 :
plt.figure() #Instantier l'objet figure 

plt.plot(x_array,f_array,label="f(x)") #Tracer y en fonction de x
plt.plot(y_array,g_array,label="g(y)") #Tracer y en fonction de x

plt.title("Graphe question 7") #Mettre un titre (optionel)
plt.xlabel("x") #Nommer l'axe des x (optionel)
plt.ylabel("f(x)") #Nommer  l'axe des y (optionel)
plt.xlim(0,100) #fixer la limite d'affichage sur x (optionel)
plt.legend()

plt.show() #Afficher la résultante

# Question 8 :
def integral_trapez(x,y):
    integrale = 0
    for i in range(len(x)-1):
        integrale += (y[i]+y[i+1])*(x[i+1]-x[i])/2 # Expression de la somme en suivant la formule (b-a)/2*(f(a)+f(b))
    return integrale

print("===> Question 8 <===")
print(f"L'intégral de f(x) vaut {integral_trapez(x_array,f_array)}")
print(f"L'intégral de g(y) vaut {integral_trapez(y_array,g_array)}")

# Question 9 :
print("===> Question 9 <===")
print(f"Pour f(x) : max={f_array.max()},min={f_array.min()},mean={f_array.mean()}")
print(f"Pour g(y) : max={g_array.max()},min={g_array.min()},mean={g_array.mean()}")
```
